```
import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
g = nx.from_pandas_edgelist(df, source='actor1surname', target='actor2surname')
```

```
print(nx.info(g))
```

```
people = list(df.actor1surname.unique())
people
```

```
plt.figure(figsize=(20,20)).set_facecolor("#191925")

plt.axis('off')
layout = nx.spring_layout(g,iterations=100)

popular_people = [person for person in people if g.degree(person) > 25]

nx.draw_networkx_nodes(g, pos=layout, node_color='#FF005F', node_size=50, label=True, alpha=0.8)
nx.draw_networkx_nodes(g, pos=layout, nodelist=popular_people, node_color='#119DA4', node_size=200, label=True, alpha=0.8)

nx.draw_networkx_edges(g, layout, edge_color='#119DA4', style='dashed', arrows=True, alpha=0.5)
nx.draw_networkx_labels(g, layout, font_color='white', font_size=8, font_weight='bold')
plt.title('15th Century Florentine Marriage Network', color='white')
```
